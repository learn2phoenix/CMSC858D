```python
test_sparse_data = pd.read_csv('test_sparse.csv')
```

```python
def plot_data(x, y, ax, title, xlabel, ylabel):
    sns.lineplot(x, y, marker='o', markers=True, ax=ax)
    ax.set_title(title, fontsize=15)
    ax.set_ylabel(ylabel, fontsize=15)
    ax.set_xlabel(xlabel, fontsize=15)
```

```python
for col in test_sparse_data.columns:
    if col not in ['Size', 'Sparsity']:
        axis_list = [[0,0], [0,1], [0,2],[0,3],[1,0],[1,1],[1,2],[1,3]]

        plt.rcParams['patch.linewidth'] = 1
        # plt.rcParams['figure.autolayout'] = True
        plt.rcParams['font.family'] = 'serif'
        plt.rcParams['font.serif'] = ['Computer Modern']
        # plt.rcParams['font.serif'] = ['Times New Roman']
        plt.rcParams['text.usetex'] = True

        sns.set_style('white')
        plt.rcParams['patch.edgecolor'] = 'none'
        fig, axes = plt.subplots(2,4, figsize=(14, 5), sharex=True, sharey=True)
        plt.subplots_adjust(bottom=.0, left=0.0,wspace=.1,hspace=0.2)
        for i, sparsity in enumerate(sorted(test_sparse_data['Sparsity'].unique())):
            sub_data = test_sparse_data[test_sparse_data['Sparsity'] == sparsity]
            plot_data(sub_data['Size'].tolist(), sub_data[col].tolist(), axes[axis_list[i][0]][axis_list[i][1]],
                     f'Sparsity:{sparsity}', 'Size of bitvector (log scale)', 'Time in ms')
        axes[0][0].set_xscale('log')
        fig.suptitle(f'Average {".".join(col.split("_"))} Times', fontsize=20, x=.45)
```

```python
for col in test_sparse_data.columns:
    if col not in ['Size', 'Sparsity']:
        axis_list = [[0,0], [0,1],[1,0],[1,1]]

        plt.rcParams['patch.linewidth'] = 1
        # plt.rcParams['figure.autolayout'] = True
        plt.rcParams['font.family'] = 'serif'
        plt.rcParams['font.serif'] = ['Computer Modern']
        # plt.rcParams['font.serif'] = ['Times New Roman']
        plt.rcParams['text.usetex'] = True

        sns.set_style('white')
        plt.rcParams['patch.edgecolor'] = 'none'
        fig, axes = plt.subplots(2,2, figsize=(8, 5), sharex=True, sharey=True)
        plt.subplots_adjust(bottom=.0, left=0.0,wspace=.1,hspace=0.2)
        for i, size in enumerate(sorted(test_sparse_data['Size'].unique())):
            sub_data = test_sparse_data[test_sparse_data['Size'] == size]
            plot_data(sub_data['Sparsity'].tolist(), sub_data[col].tolist(), axes[axis_list[i][0]][axis_list[i][1]],
                     f'Size:{size}', 'Sparsity of bitvector (log scale)', 'Time in ms')
        axes[0][0].set_xscale('log')
        fig.suptitle(f'Average {".".join(col.split("_"))} Times', fontsize=20, x=.45)
```
